```python
import gradio as gr
from transformers import pipeline, Conversation
import torch
```

```python
chatbot = pipeline(task="conversational", model="facebook/blenderbot-400M-distill")
```

```python
# Define a function that generates responses from the chatbot
def generate_response(input_text):
    conversation = Conversation(text=input_text)
    response = chatbot(conversation, )

    return response.generated_responses[-1]
```

```python
interface = gr.Interface(
    fn=generate_response,
    inputs=[
        gr.Textbox(lines=2, label="User Input", placeholder="Enter your message here...")
    ],
    outputs=[
        gr.Textbox(lines=2, label="LLM output")
    ],
    title="Dumb Bot",
    description="A chatbot using Hugging Face Blenderbot model having 400 million parameters.",
    examples=["tell me about the burj khalifa"]
)
```

```python
interface.launch()
```

```python
gr.close_all()
```
